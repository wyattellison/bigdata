````python
#Run this cell, expect an error
time = 5
print("It is currently " + time + " PM")
````

Oh no! We got ```TypeError: can only concatenate str (not "int") to str```. Python can't concatenate strings to integers, so it threw an error. One way to get around this is:

````python
print("It is currently " + str(time) + " PM")
````

This is called casting. Using the ```str()``` function *casts* the integer represented by ```time``` to a string, which can be concatenated to other strings. Mr. Manzo thinks that Java is better because of this, but there's something called ```fstrings``` which are very useful... Here's how to use an ```fstring```:

````python
#Run this cell
print(f"It is currently {time} PM")
````

So by putting ```f``` before the string, you are able to wrap the name of your variable in curly brackets and print it without having to concatenate strings. Python 1, Java 0.

# VARIABLES

There are a few different kinds of variables in Python with which we'll now concern ourselves. Python is an object-oriented programming language, so for now you can think of your variables as objects, and we'll add a few more later.

### Integers

Integers (int) are are whole numbers. If you have any more questions about that, ask your math teacher. In Python3, ints have no size limit until the memory of the computer is used. Define an int by setting a variable equal to any whole number:
```
my_int = 5
your_int = 4
their_int = 3141592682939101938585030239474575891928274859
```

### Strings

Strings (str) are like words. Define a string with quotation marks or apostrophes:
```
my_string = "hello world!"
your_string = 'hello world.'
```
Or you can define a multiline string with triple quotation marks or triple apostrophes:
```
their_string = '''this is a haiku
mr manzo can write code
is it spaghetti?'''
```
~(Yes, Manzo's code is spaghetti)~


In Python, like other languages, strings are stored as a list of unicode characters. This means we can iterate them:


````python
#Run this cell
my_string = 'hello world'
print(my_string[0])
#Play around with the integer in the square brackets. Remember, Python indexing starts at 0...
````

### Floats
Floats are floating point decimals, which is just a way of saying they're numbers that can have decimals (but they can also be whole numbers). Define a float by setting a variable equal to any number, but there has to be a decimal:
```
my_float = 3.14
your_float = 3.1415
their_float = 3.
```
Notice how ```their_float``` is a whole number, but it's not an int since they put a decimal point after 3

### Lists
Lists are... lists. Much more on lists later, but for now, know that a list can store other variables inside of it (including other lists... hmm, interesting...). Lists have a certain order to them, and can be iterated (like the strings example from earlier, since Python treats strings as a list of characters). Python lists need not have the same type of variable for the whole list. Define a list using square brackets and commas as delimiters:
```
my_list = ['eggs', 'bananas', 'apples', 'milk']
your_list = ['caviar','mineral water', 'filet mignon']
their_list = [6, 92.75, 'noble and strong',[2,5]]
```

````python
#Run this cell
my_list = ['eggs', 'bananas', 'apples', 'milk']
print(my_list[0])
#Play around with the integer in the square brackets.
````

Since ```my_list``` is a list of strings, and strings are a list of characters... Doesn't that mean we can iterate the items in ```my_list```? Let's try that:

````python
#Run this cell
print(my_list[0][0])
#Play around with the different indices (integers in the square brackets)
````

And just a reminder how modulo works it's the remainder after dividing.
```
5 % 3 = 2
16 % 5 = 1
10 % 2 = 0
```

Also, here's a little trick for exponents.  You can definitely do simple exponents like
```
3 ** 2 = 9
10 ** 3 = 1000
```

You can also do roots, but you have to use the decimal exponent or use numbers with a decimal place.
```
4 ** 0.5 = 2.0

27 ** 0.333 = about 3

16 ** 1/2 = 8.0
```
This is surprising, right?  Why does this happen?  Python reads this as 16 ** 1 and then divided by 2. Remember your PEMDAS...

Notice how there's two kinds of dividing operators. The normal divide (also called floating point division) with ```/``` and integer division with ```//```. Floating point division is pretty standard division, but integer division only returns the quotient as an integer by cutting off all the decimal places. It's kind of like the opposite of the modulus.

```
16 ** (1/2) = 4.0
16 ** (1//2) = 1
```
The first line is 16 raised to .5 (which is a float, so it returns a float), and the second line is 16 raised to the rounded down integer of .5, which is 0, and anything raised to 0 is 1. Since it's integer division and exponents, the result is an integer.
